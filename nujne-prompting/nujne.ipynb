```python
import json
import re

# File paths
input_path = r"D:\FAKS\MAG\LET2\FAKS\NLP\S\ul-fri-nlp-course-project-2024-2025-jazbeci\nujne-prompting\nujne_dvojne_simplified.json"
output_path = r"D:\FAKS\MAG\LET2\FAKS\NLP\S\ul-fri-nlp-course-project-2024-2025-jazbeci\nujne-prompting\nujne_extracted.json"

# Load data
with open(input_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# Function to extract timestamp and content
def extract_timestamp_and_content(text):
    # Extract timestamp (date and time)
    date_time_match = re.search(r'(\d{2}\.\s\d{2}\.\s\d{4})\s+(\d{2}\.\d{2})', text)
    if date_time_match:
        date = date_time_match.group(1)
        time = date_time_match.group(2)
        timestamp = f"{date} {time}"
    else:
        timestamp = "N/A"
    
    # Extract content (after "Nujna prometna informacija.")
    content_match = re.search(r'Nujna prometna informacija\.?\n\n(.*?)$', text, re.DOTALL)
    if content_match:
        content = content_match.group(1).strip()
    else:
        # Alternative pattern for newer formats
        content_match = re.search(r'NUJNA PROMETNA INFORMACIJA\s*\n\n(.*?)$', text, re.DOTALL)
        if content_match:
            content = content_match.group(1).strip()
        else:
            content = "Content not found"
    
    return {
        "timestamp": timestamp,
        "content": content
    }

# Process each entry
extracted_data = [extract_timestamp_and_content(entry) for entry in data]

# Save result
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(extracted_data, f, ensure_ascii=False, indent=4)

# Print the result
print(json.dumps(extracted_data, ensure_ascii=False, indent=4))

print(f"Processed {len(extracted_data)} entries and saved to {output_path}")
```

```python
import json
from docx import Document
import os

# File paths
input_path = r"D:\FAKS\MAG\LET2\FAKS\NLP\S\ul-fri-nlp-course-project-2024-2025-jazbeci\nujne-prompting\nujne_extracted.json"
output_path = r"D:\FAKS\MAG\LET2\FAKS\NLP\S\ul-fri-nlp-course-project-2024-2025-jazbeci\nujne-prompting\nujne_traffic_alerts.docx"

# Load the extracted data
with open(input_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# Create a new Word document
doc = Document()
doc.add_heading('Traffic Alerts', 0)

# Add each entry as a paragraph
for entry in data:
    timestamp = entry.get("timestamp", "N/A")
    content = entry.get("content", "Content not found")
    
    # Combine timestamp and content into a single line
    combined_text = f"{timestamp} - {content}"
    doc.add_paragraph(combined_text)

# Save the document
doc.save(output_path)

print(f"Created Word document with {len(data)} traffic alerts at: {output_path}")
```
